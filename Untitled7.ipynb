```python
#series - column in table,1D array holdind data of any type

import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a)

print(myvar)


#labels-If nothing else is specified, the values are labeled with their index number.

print(myvar[0])
print(myvar[2])
```

```python
#create labels

import pandas as pd

a = [1, 7, 2]

myvar = pd.Series(a, index = ["x", "y", "z"])

print(myvar)

print(myvar["y"])
```

```python
#Key/Value Objects as Series

import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories)

print(myvar)
```

```python
import pandas as pd

calories = {"day1": 420, "day2": 380, "day3": 390}

myvar = pd.Series(calories, index = ["day1", "day2"])

print(myvar)
```

```python
#Dataframes - whole table

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

myvar = pd.DataFrame(data)

print(myvar)
```

```python
import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

#load data into a DataFrame object:
df = pd.DataFrame(data)

print(df) 
```

```python
#Pandas use the loc attribute to return one or more specified row(s)

#refer to the row index:
print(df.loc[0])
```

```python
#use a list of indexes:
print(df.loc[[0, 1]]) #When using [], the result is a Pandas DataFrame.
```

```python
#With the index argument, you can name your own indexes.

import pandas as pd

data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(df) 

#refer to the named index:
print(df.loc["day2"])
```
