```python
#Scope

#local scope
def myfunc():
  x = 300
  print(x)

myfunc()
```

```python
def myfunc():
  x = 300
  def myinnerfunc():
    print(x)
  myinnerfunc()

myfunc()
```

```python
#global scope
x = 300

def myfunc():
  print(x)

myfunc()

print(x)

```

```python
x = 300

def myfunc():
  x = 200
  print(x)

myfunc()

print(x)
```

```python
#Global Keyword - keyword makes the variable global.
def myfunc():
  global x
  x = 300

myfunc()

print(x)
```

```python
x = 300

def myfunc():
  global x
  x = 200

myfunc()

print(x)
```

```python
#Nonlocal Keyword - keyword is used to work with variables inside nested functions.

def myfunc1():
  x = "Jane"
  def myfunc2():
    nonlocal x
    x = "hello"
  myfunc2()
  return x

print(myfunc1())
```

```python
x = "global"

def outer():
  x = "enclosing"
  def inner():
    x = "local"
    print("Inner:", x)
  inner()
  print("Outer:", x)

outer()
print("Global:", x)
```
